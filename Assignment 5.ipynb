```python
#1. Guess a number is a game that prompts a player to guess a number between 0 and 9, which
#  is randomly generated by the system. When the input given by the user matches the
#  number generated by the system then the user wins. The game should go as follows:

# Guess the number: 5
# Sorry, try again
# Guess the number: 3
# Sorry, try again
# Guess the number: 8
# You got it right! Congo!

import random
i=0
randomm=random.randint(0,9)
while i<3:
    num=eval(input("guess the number :"))
    if num==randomm:
        print("you got it right! congo!")
        break
    elif num!=randomm:
        print("sorry,try again")
    i=i+1
        
```
