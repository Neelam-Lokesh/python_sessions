# Local variables & Global variables

- local variables provided inside the function
- Global variaBLES provided outside the function
- LV are inside function
- LV values can not use out side the function until unless you return those values
- if we want to convert local var to global var we must add Global keyword 



```python
# local variables

def addition():
    
    n1=10
    n2=20
    add=n1+n2
    print(" addition of {} and {} is {}".format(n1,n2,add))
    
    
addition()
```

```python
# global variables

n1=10
n2=20
def addition():                                                     #no need to provide inside brackets
    add=n1+n2
    print(" addition of {} and {} is {}".format(n1,n2,add))
    
    
addition()
```
